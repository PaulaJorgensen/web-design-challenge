```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file = "cities.csv"

# Read Purchasing File and store into Pandas data frame
data = pd.read_csv(file)

data
```

```python

data_df = pd.DataFrame({'City ID':data['Unnamed: 0'], 'City':data['city'], 'Temp': data['temp'], 'Max Temp': data['max temp'],
                       'Humidity':data['humidity'], 'Cloudiness':data['cloudiness'], 'Wind Speed':data['wind speed'],
                        'LAT':data['lat'], 'LON':data['lon'], 'City Country':data['city country'], "City Datetime":data['city datetime']})
data_df.head()
```

```python
#render dataframe as html
html = data_df.to_html()

```

```python
#write html to file
text_file = open("DataRaw.html", "w")
text_file.write(html)
text_file.close()
```
